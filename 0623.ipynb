```python
#구구단 2단부터 9단까지 출력하는 프로그램
for i in range(2,10):# for 반복문 이용 - 2~9 시퀀스 data 입력
  for j in range(2,10): # 2~9 반복하는 반복문, 변수는 j에 값 저장
    print(f'{i} X {j} = {i*j}',end='\t') #출력기능 - 구구단 내용 1회 출력, 줄바꿈 제거
  else: #반복문이 완전하게 정지시 동작
   print() #출력기능 - 구구단 1단이 끝나면 줄바꿈용 출력
else: #반복문이 완전하게 정지시 동작
  print() #줄바꿈
  print("프로그램이 종료되었습니다.") #프로그램의 완전한 종료 확인용 출력
```
