<b>Q3 난수생성(1~100)이요 100명 학생 점수 만들고 조건에 맞게 출력<br>
1. 100명 평균 50점 이상이면 전원통과<br>
2. 100명 모두가 50점 이상이면 학년1등반 출력<br>
3. 1명이라도 100점이 있다면 전교1등이 있는 반 출력

```python
np.random.randint(0,101,10)
```

```python
import numpy as np
x=np.random.randint(10,101,100)
if np.mean(x)>=50:
  print("전원통과")
if np.all(x>=50):
  print("학년 1등")
if np.any(x==100):
  print("전교1등이 있는 반")
```
