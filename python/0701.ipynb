<b> pickle

```python
import pickle
data={"key1":"data","key2":"data2"}
pickle.dump(data,open("d_data.p",'wb'))  #dump(무엇을,어디에(통로를 열어서 진짜이름)) #쓰기위해 보내는거이기때문에 w이고 딕셔너리 이므로 binary모드로 열어야함
#피클은 자체적으로 수정 불가능
#객체로 받았기 때문에 객체에서 수정 가능
```

```python
ck=pickle.load(open("d_data.p",'rb'))
```

```python
ck
```

```python
import json 
json.dump(ck,open("d_data.json",'w'))#json은 텍스트로 취급한다.
```

```python
ck2=json.load(open("d_data.json",'r'))
ck2
```

```python
try:
  open("d1.txt",'r') #예외가 발생할수 있는 범위
  print("남은내용")
except: #except 조건이 참
  print('예외발생')
else:
  print("예외가 발생하지 않았습니다.")
```

```python
try:
  open("d.txt",'r') #예외가 발생할수 있는 범위
  print("남은내용")
except: #except 조건이 거짓
  print('예외발생')
else: #예외 미발생 시 실행되는 블록
  print("예외가 발생하지 않았습니다.")
```

```python
try:
  open("d1.txt",'r') #예외가 발생할수 있는 범위
  print("남은내용")
except IndexError: #인덱스 에러라는 클래스로 만들어진 예외
  print("인덱스 예외")
except: 
  print('예외발생')
else: #예외 미발생 시 실행되는 블록
  print("예외가 발생하지 않았습니다.")
```

```python
try:
  open("d1.txt",'r') #예외가 발생할수 있는 범위
  print("남은내용")
except Exception: # Exception 모든것을 포함하는 거임
  print('예외발생')
except IndexError: #인덱스 에러라는 클래스로 만들어진 예외
  print("인덱스 예외")
else: #예외 미발생 시 실행되는 블록
  print("예외가 발생하지 않았습니다.")
```

```python
try:
  #open("d1.txt",'r') #예외가 발생할수 있는 범위
  print("남은내용")
  l=[]
  print(l[0])
except Exception: # Exception 모든것을 포함하는 거임
  print('예외발생')
except IndexError: #인덱스 에러라는 클래스로 만들어진 예외
  print("인덱스 예외")
else: #예외 미발생 시 실행되는 블록
  print("예외가 발생하지 않았습니다.")
```

```python
try:
  #open("d1.txt",'r') #예외가 발생할수 있는 범위
  print("남은내용")
  l=[]
  print(l[0])
except IndexError: #인덱스 에러라는 클래스로 만들어진 예외
  print("인덱스 예외")
except Exception: # Exception 모든것을 포함하는 거임
  print('예외발생')
else: #예외 미발생 시 실행되는 블록
  print("예외가 발생하지 않았습니다.")
```

```python
try:
  f=open("d.txt",'r') #가정->절대 예외가 발생하지 않는 코드
  print("남은내용")
  l=[]
  print(l[0])
except IndexError: #인덱스 에러라는 클래스로 만들어진 예외
  print("인덱스 예외")
except Exception: # Exception 모든것을 포함하는 거임
  print('예외발생')
else: #예외 미발생 시 실행되는 블록
  print("예외가 발생하지 않았습니다.")
finally: #반드시 동작하는 블록
  #자원정리의 목적, 반드시 정리한다
   f.close() #예외가 발생해도 닫히고, 발생하지 않아도 닫힌다.
```
