```python
#1 100개의 정수를 저장할 수 있는 배열을 선언하고 정수를 차례대로 입력받다가 0이 입력되면 입력을 종료한다. 0을 제외하고 입력받은 수를 가장 나중에 입력된 정수부터 차례대로 출력하는 프로그램
# 다중입력과 단일 입력 두가지를 동시에 제공하는 프로그램
l2=[]
for i in range(100):
  in_data=list(map(int,input("정수입력 (구분자 ' ' ):").split())) #정수로 입력받은 데이터들을 리스트로 표현
  if len(in_data)>1:# 다중입력을 확인,입력된 갯수 여러개, 즉 여러개가 입력되었다.
    if 0 in in_data: # in_data 안에 0의 값이 있는지 묻는 조건문
      l2.extend(in_data[:in_data.index(0)])# 인덱스가 0인 원소 전까지 하나씩 꺼내온다
      break #반복루프, 반복에 관한 블럭 한번 나간다 ->19번으로 이동
    else:
      l2.extend(in_data)
      if len(l2)>=100:
        l2=l2[:100] #100개까지만 출력
        break #반복루프, 반복에 관한 블럭 한번 나간다 ->19번으로 이동
  else: # 단일 입력
    ck=in_data[0] #0이 입력되었는지 확인 하는 변수 선언
    if ck==0:
      break
    l2.append(ck) #ck에 원소를 넣어준다
  if len(l2)>=100:
        l2=l2[:100] #100개까지만 출력
        break #반복루프, 반복에 관한 블럭 한번 나간다 ->19번으로 이동
#19번줄
print(l2[::-1]) #역순으로 출력
    
```

```python
#3 영문 대문자를 입력 받다가 대문자 이외의 문자가 입력되면 입력을 중단하고, 1번 이상 입력된 문자와 그 문자의 개수를 출력하는 프로그램
#한글자만 출력
dic={}
while True:
  in_data=input("대문자 입력: ")[0] #한글자만 출력
  if not ('A'<=in_data<='Z'):
    break
  if in_data in dic.keys():
    dic[in_data]+=1 #기존에 있는 값이면 +1
  else:
    dic[in_data]=1  # 처음 입력된 값이면 1로 설정
for key,v in dic.items():
  print(f'{key}:{v}')
```

```python
l=[0 for _ in range(26)] #파이썬의 리스트는 존재하지는 않는 공간에 저장할수 없다. 따라서 리스트의 빈 공간을 만든다.
while True: #무한루프 생성
  in_data=input("대문자 입력: ")[0] #한글자만 출력
  if not ('A'<=in_data<='Z'): #입력값이 대문자에 속하는지 확인
    break
  l[ord(in_data)-ord('A')]+=1
for i in range(len(l)):
  if l[i]>0:
    print(chr(ord('A')+i),':',l[i])
```

```python
#5 정수형 N을 입력받아 다음과 같은 출력을 하는 프로그램
dx=[1,0,-1,0] #저차원 방향성
dy=[0,1,0,-1] #고차원 방향성
d=0 #방향 인덱스
x=0
y=0
n=int(input("정수 n입력:"))
a=[[0 for i in range(n)] for i in range(n)] #0으로 가득 찬 1차원 을 원소로 가지는 2차원 리스트 생성
for i in range(1,n*n+1):
  a[y][x]=i #이름에서 가까우면 고차원, 멀면 저차원 리스트에 data 입력
  yy=y+dy[d] #다음 y좌표 예상,변화된 y
  xx=x+dx[d] #다음 x좌표 예상, 변화된 x
  if (0<=yy<n) and (0<=xx<n): #범위 확인
    if a[yy][xx]==0:          #값유무 확인 
      y=yy
      x=xx
    else:                       #방향전환 조건
      d=(d+1)%4 #0~3까지만 나옴 
      x+=dx[d]
      y+=dy[d]
  else:                       #방향전환 조건
    d=(d+1)%4 #0~3까지만 나옴
    x+=dx[d]
    y+=dy[d]
for a1 in a:
  for i in a1:
    print(i,end='\t')
  print()
```
