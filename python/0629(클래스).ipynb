```python
class 클래스:
  클래스_필드='클래스필드_호출'
  #생성자
  def __init__(self,인스턴스입력='인스턴스필드_호출'): #생성자 과정에서 인스턴스필드 만든다. 왜냐하면 인스턴스 필드는 인스턴스가 생성될때 만들어지기 때문
    self.인스턴스_필드=인스턴스입력

    #인스턴스메소드->인스턴스필드
  def 인스턴스메소드_인스턴스필드(self):
    print(f"인스턴스_메소드->인스턴스_필드:{self.인스턴스_필드}") #인스턴스 안에 있기 때문에 self.을 붙여야한다
    
    #인스턴스메소드->클래스필드
  def 인스턴스메소드_클래스필드(self):
        print(f'인스턴스_메소드->클래스_필드:{self.클래스_필드}')

    #클래스메소드->클래스필드
  @classmethod
  def 클래스메소드_클래스필드(cls): #cls는 클래스를 지칭하는 메소드
    print(f'클래스_메소드->클래스_필드:{cls.클래스_필드}')

    #클래스메소드->인스턴스필드
  @classmethod
  def 클래스메소드_인스턴스필드(cls): #cls는 클래스를 지칭하는 메소드
    #print(f'클래스_메소드->인스턴스_필드:{cls.인스턴스_필드}')
    print("접근불가")

#클래스 출발

#클래스 -O->클래스_메소드-O->클래스필드
#접근가능
클래스.클래스메소드_클래스필드()

#클래스-O->클래스_메소드-X->인스턴스필드
#접근불가(클래스 -X-> 인스턴스)
클래스.클래스메소드_인스턴스필드()

#클래스-X->인스턴스_메소드-O->클래스필드
#접근불가(클래스 -X-> 인스턴스)
#클래스.인스턴스메소드_클래스필드()

#클래스-X->인스턴스_메소드-O->인스턴스필드
#접근불가(클래스 -X-> 인스턴스)
#클래스.인스턴스메소드_인스턴스필드()


#인스턴스 출발

인스턴스=클래스()

#인스턴스-O->클래스_메소드-O->클래스필드
#접근가능
인스턴스.클래스메소드_클래스필드()

#인스턴스-O->클래스_메소드-X->인스턴스필드
#접근불가
인스턴스.클래스메소드_인스턴스필드()

#인스턴스-O->인스턴스_메소드-O->클래스필드
#접근가능
인스턴스.인스턴스메소드_클래스필드()

#인스턴스-O->인스턴스_메소드-O->인스턴스필드
#접근가능
인스턴스.인스턴스메소드_인스턴스필드()
```
