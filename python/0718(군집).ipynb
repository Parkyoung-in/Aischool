<b><h3>K_Means

```python
f_2D=f.reshape(-1,100*100)
```

```python
from sklearn.cluster import KMeans #cluster 군집
km=KMeans(n_clusters=3,random_state=42)
km.fit(f_2D)
```

```python
np.unique(km.labels_,return_counts=True)
```

```python
def d_f(arr,ratio=1):
  n=len(arr)
  rows=int(np.ceil(n/10)) 
  cols = n if rows<2 else 10
  fg,ax = plt.subplots(rows,cols,squeeze=False)
  for i in range(rows):
    for j in range(cols):
      if i *10+j<n:
        ax[i,j].imshow(arr[i*10+j],cmap='gray_r')
      ax[i,j].axis('off')
  plt.show()
```

```python
d_f(f[km.labels_==2])
```

```python
def d_f(arr,ratio=1):
  n=len(arr)
  rows=int(np.ceil(n/10))
  cols = n if rows<2 else 10
  fg,ax = plt.subplots(rows,cols,figsize=(cols*ratio,rows*ratio),squeeze=False)
  for i in range(rows):
    for j in range(cols):
      if i *10+j<n:
        ax[i,j].imshow(arr[i*10+j],cmap='gray_r')
      ax[i,j].axis('off')
  plt.show()
```

<b><h3>최적의 K

```python
ck=[]
for k in range(2,5):
  km=KMeans(n_clusters=k,random_state=42)
  km.fit(f_2D)
  ck.append(km.inertia_)
plt.plot(range(2,5),ck)
plt.show()
```
