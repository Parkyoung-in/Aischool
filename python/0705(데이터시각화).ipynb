```python
import numpy as np
c_Y=[np.cos(i) for i in X] #코사인으로 만들어진 Y
s_Y=[np.sin(i) for i in X] #사인으로 만들어진 Y
plt.plot(s_Y)
plt.plot(c_Y)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(c_Y)
ax.set(title='cos_g',xlabel='n',ylabel='y')
plt.show()
```

```python
fig=plt.figure() #그래프의 정보는 축에 넣는다
fig.set_size_inches(10,10)
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
ax1.plot(c_Y)
ax2.plot(s_Y)
plt.show()
```

```python
fig,ax=plt.subplots(2,2)
x=np.linspace(-1,1,100)
y_1=np.sin(x)
y_2=np.cos(x)
y_3=np.tan(x)
y_4=np.exp(x)
ax[0,0].plot(x,y_1)
ax[0,1].plot(x,y_2)
ax[1,0].plot(x,y_3)
ax[1,1].plot(x,y_4)
plt.show() #원래는 show를 해야 보인다. #show를 하면 초기화 효과도 있다. -> 반복문 쓸수 있다.
```
