<b>조인

```python
d1={
    'c_id':['1','2','3','4','5'],
   'data':[10,20,30,40,50]
}
df_l=DataFrame(d1)
df_l
```

```python
d2={
    'c_id':['3','4','5','6','7'],
   'data':[10,20,30,40,50], 
    'data1':[100,200,300,400,500]
}
df_r=DataFrame(d2)
df_r
```

```python
pd.merge(left=df_l,right=df_r,how='inner',on='c_id') #in:어떤 조인?  #on:기준 #내부조인
```

```python
pd.merge(left=df_l,right=df_r,how='left',on='c_id')  #왼쪽조인 #왼쪽은 1,2,3,4,5가 있고 오른쪽은 3,4,5,6,7만 있으므로 없는 곳은 공간을 만들되 공백으로 둔다
```

```python
pd.merge(left=df_l,right=df_r,how='right',on='c_id') #오른쪽 조인
```

```python
pd.merge(left=df_l,right=df_r,how='outer',on='c_id') #외부조인
```
