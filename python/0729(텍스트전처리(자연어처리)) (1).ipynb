<b><h3>정수인코딩

```python
from nltk.tokenize import sent_tokenize #문장 토큰화
from nltk.tokenize import word_tokenize #단어 토큰화
from nltk.corpus import stopwords #불용어
```

```python
text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
#코퍼스
```

```python
문장토큰화_된_data=sent_tokenize(text)
문장토큰화_된_data
```

```python
#단어 토큰화는 단순히 단어로 쪼개는 것이 아니라 의미 있는 단어와 없는 단어 정리
#데이터로 넣으려는 voacbulary 생성
#영어는 대소문자 통합 -> 소문자로 통합하면됨(lower())
```

```python
nltk.download('stopwords')
```

```python
단어모음={}
t_data=[]
불용성단어=set(stopwords.words('english')) #중복된 단어는 삭제하기 위해 set을 통해 생성 #set : 원소 중복 불가
```

```python
for 문장 in 문장토큰화_된_data:
  단어토큰화_된_data=word_tokenize(문장)
  l=[]
  for 단어 in 단어토큰화_된_data:
    소문자화된_단어=단어.lower()
    if 소문자화된_단어 not in 불용성단어:
      if len(소문자화된_단어)>2:
        l.append(소문자화된_단어)
        if 소문자화된_단어 not in 단어모음: #단어 등장의 빈도수 확인 #단어모음 = 딕셔너리(키값)
          단어모음[소문자화된_단어]=0
        단어모음[소문자화된_단어]+=1
  t_data.append(l)
t_data
```

```python
단어모음 #빈도수 확인
```

```python
 정렬된_단어모음=sorted(단어모음.items(),key=lambda x:x[1],reverse=True) #sorted를 쓰는 이유는 복합데이터에서 원하는 기준으로 줄세우기 위해서 사용 
 #빈도수를 키값으로 정렬하기위해 #내림차순 하기 위해 리버스
 정렬된_단어모음
```

```python
단어_인덱스={}
i=0
for (단어,빈도수) in 정렬된_단어모음: #i만 하면 키값만 전달
  if 빈도수>1:
    i+=1
    단어_인덱스[단어]=i
단어_인덱스
```

```python
살릴단어수=4
제거_data=[단어 for 단어,i in 단어_인덱스.items() if i > 살릴단어수]
for i in 제거_data:
  del 단어_인덱스[i]
단어_인덱스
```

```python
단어_인덱스['oov'] = len(단어_인덱스)+1 #oov : out of vocabulary #oov를 단어로 취급할때 #딕셔너리라서 현재 담겨있는 단어중 가장 마지막 위치 #전체적인 단어에 대한 길이 확인 
#딕셔너리는 번호를 가지고 있으므로 +1을 더 해야한다
단어_인덱스
```

단어인덱스(vocabulary,단어모음집)을 반드시 기억하고 있어야한다. 그래야 oov가 문자로 입력받아도 쓸수 있다

```python
t_data
```

```python
#정수인코딩
ec_data=[]
for 문장 in t_data:
  ec_d=[]
  for 단어 in 문장:
    try : #예외처리 하는 이유 : 텍스트라서
      ec_d.append(단어_인덱스[단어])
    except KeyError:
      ec_d.append(단어_인덱스['oov'])
  ec_data.append(ec_d) #for문이 끝났으므로 한 문장이 완성 됨
ec_data
```

<b><h3>데이터 수집(코퍼스 수집)

```python
text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
```

<b><h3>토큰화

```python
from nltk.tokenize import sent_tokenize #문장 토큰화
from nltk.tokenize import word_tokenize #단어 토큰화
from nltk.corpus import stopwords
```

```python
문장토큰화_된_data=sent_tokenize(text)
단어모음={}
t_data=[]
불용성단어=set(stopwords.words('english'))
for 문장 in 문장토큰화_된_data:
  단어토큰화_된_data=word_tokenize(문장)
  l=[]
  for 단어 in 단어토큰화_된_data:
    소문자화된_단어=단어.lower()
    if 소문자화된_단어 not in 불용성단어:
      if len(소문자화된_단어)>2:
        l.append(소문자화된_단어)
        if 소문자화된_단어 not in 단어모음: #단어 등장의 빈도수 확인 #단어모음 = 딕셔너리(키값)
          단어모음[소문자화된_단어]=0
        단어모음[소문자화된_단어]+=1
  t_data.append(l)
t_data
```

<b><h3>정수인코딩

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tk=Tokenizer()
tk.fit_on_texts(t_data)
ec_data=tk.texts_to_sequences(t_data)
ec_data
```
