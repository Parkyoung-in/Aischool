```python
g=sns.FacetGrid(tips,col='time',row='sex') # 가로 : 시간 세로 : 성별
#map(입력 함수,데이터구조) 데이터구조를 입력함수를 통해 하나씩 하나씩 적용해준다  
g.map(sns.scatterplot,'total_bill','tip') #산점도그래프(토탈빌과 팁을 나타냄) #산점도는 두가지 표현
```

```python
g=sns.FacetGrid(tips,col='time',row='sex')
#map(입력 함수,데이터구조) 데이터구조를 입력함수를 통해 하나씩 하나씩 적용해준다  
g.map(sns.histplot,'total_bill')  #히스토그램은 하나만 표현
```
