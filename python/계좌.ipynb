```python
ui='''----------------------------------------------------------------------------
1.계좌생성 | 2.계좌목록 | 3.예금 | 4.출금 | 5.송금 | 6.내역 조회 | 7.종료
----------------------------------------------------------------------------
선택> '''
```

```python
고객정보=[]
입출금기록=[]
while True:
  n=int(input(ui))
  #1. 계좌생성
  if n==1:
    print("--------------\n계좌생성\n--------------")
    계좌번호=input("계좌번호: ")
    고객정보.append(계좌번호)
    계좌주=input("계좌주: ")
    고객정보.append(계좌주)
    초기입금액=int(input("초기입금액: "))
    고객정보.append(초기입금액)
    입출금기록.append(초기입금액)
    print("결과: 계좌가 생성되었습니다.")
  elif n==2:
    print("--------------\n계좌목록\n--------------")
    print(고객정보)
  elif n==3:
    print("--------------\n예금\n--------------")
    계좌번호=input("계좌번호: ")
    예금액=int(input("예금액: "))
    고객정보[2]=고객정보[2]+예금액
    입출금기록.append(예금액)
    print("예금이 성공되었습니다.")
  elif n==4:
    print("--------------\n출금\n--------------")
    계좌번호=input("계좌번호: ")
    출금액=int(input("출금액: "))
    if 출금액 > 고객정보[2]:
      print("잔액이 부족합니다.")
    else:
      print("결과 : 출금이 성공되었습니다.")
      고객정보[2]=고객정보[2]-출금액
      입출금기록.append(-1*출금액)
  elif n==5:
    print("--------------\n송금\n--------------")
    계좌번호=input("계좌번호: ")
    받는분계좌=input("받는사람계좌번호: ")
    송금액=int(input("송금액: "))
    if 송금액 > 고객정보[2]:
      print("잔액이 부족합니다.")
    else:
      print("결과 : 송금이 성공되었습니다.")
      고객정보[2]=고객정보[2]-송금액
      입출금기록.append(-1*송금액)
  elif n==6:
    print("--------------\n내역조회\n--------------")
    계좌번호=input("계좌번호: ")
    print(입출금기록)
  elif n==7:
    print("프로그램 종료")
    break
```
