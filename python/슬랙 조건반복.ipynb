```python
n=int(input('n:'))#입력 data개수 결정
n_l=[]#입력 data 저장장치
while True: #단일 다중 입력 장치
  입력:input('정수입력기(구분자 " "):').split()#리스트 만들어짐->원소의 갯수
  if len(입력)>1: #다중입력장치
    n_l.extend(list(map(int,입력))[:n]) #최소 2개 이상이 입력 = int로 바꿔줘서 하나씩 입력
  else: #1개의 원소가 입력될수 있게 해줌
    n_l.append(int(입력[0]))
  if len(n_l)>=n:
    n_l=n_l[:n]
    break
#5개 입력방지기
for i in n_l:#무조건 data는 n개 입력
  if i <0:
    break
```
