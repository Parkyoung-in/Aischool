```python
#dictionary인자
def f(**x): #딕셔너리 인자는 key,value 2개이므로 '*'이 2개 붙는다.
  print(x)
  x[x.keys()[0]]=10
  print(x)
f(b=2,c=3,d=4)
```

```python
#dictionary인자
def f(**x): #딕셔너리 인자는 key,value 2개이므로 '*'이 2개 붙는다.
  print(x)
  x[list(x.keys())[0]]=10
  print(x)
f(b=2,c=3,d=4)
```
