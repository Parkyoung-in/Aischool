```python
#피벗(기준) 테이블
titanic
```

```python
titanic.pivot_table(index='class',columns='sex',aggfunc='size') #세로=index 가로=행 #aggfunc='기준' : 집계펑션을 통해 정렬
```

```python
data=titanic.pivot_table(index='class',columns='sex',aggfunc='size')
```

```python
titanic.pivot_table(index='class',columns='sex',values='fare')
```

```python
sns.heatmap(data,annot=True)#세로축 : 데이터의 양
```

```python
sns.heatmap(data,annot=True,fmt='d') #세로축 : 데이터의 양 #기본값이 정수가 아니므로 정수로 설정
```

```python
flights.pivot('month','year','passengers')#추출의 목적을 위해 피벗을 쓴다 (축,축,기준이 되는 대상(vlaue값))
```

```python
t_data=flights.pivot('month','year','passengers')
```

```python
sns.heatmap(t_data)
```

```python
sns.heatmap(t_data,annot=True, fmt='d') #annot 숫자 출력 fmt=d : 정수로 표현
```

```python
sns.heatmap(t_data,annot=True, fmt='d',linewidths=1) #선을 그어 구분 선은 1로 설정
```
